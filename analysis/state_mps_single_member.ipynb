```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
```

```python
state_mps = {}

state_mps['TAS'] = {
    "Cassy O'Connor (LegCo)": np.arange(2024, 2026),
}

state_mps['WA'] = {
    "Adele Carles": np.arange(2009, 2010),
}

state_mps['NSW'] = {
    "Kobi Shetty": np.arange(2023, 2026),
    "Tamara Smith": np.arange(2015, 2026),
    "Jenny Leong": np.arange(2015, 2026),
    "Jamie Parker": np.arange(2011, 2023),
}

state_mps['SA'] = {
    "Kris Hanna": np.arange(2003, 2006),
}
# MacLaren was also an MP for 3 months during 2015

state_mps['VIC'] = {
    "Gabrielle de Vietri": np.arange(2022, 2026),
    "Tim Read": np.arange(2018, 2026),
    "Lidia Thorpe": np.arange(2017, 2018),
    "Ellen Sandell": np.arange(2014, 2026),
    "Sam Hibbins": np.arange(2014, 2024),
}
# Huong Truong replaced Colleen Hartland for much of 2018

state_mps['QLD'] = {
    "Amy MacMahon": np.arange(2020, 2024),
    "Michael Berkman": np.arange(2017, 2026),
    "Ronan Lee": np.arange(2008, 2009),
}

state_mps['NT'] = {
    "Kat McNamara": np.arange(2024, 2026),
}
```

```python
#mp_count_dict = {'year': np.arange(1972, 2023)}
mp_count_dict = {'year': np.arange(2003, 2026)}
for state, mp_dict in state_mps.items():
    years = np.concatenate(list(mp_dict.values()))
    counts, lower_bin_bounds = np.histogram(years, bins=np.arange(2002.5, 2026.5))
    mp_count_dict[state] = counts
```

```python
df = pd.DataFrame(mp_count_dict)
df = df.set_index('year')
order = ['SA', 'QLD', 'WA', 'NSW', 'VIC', 'TAS', 'NT']
df = df[order]
df
```

```python
df.plot(kind='bar', stacked=True, figsize=(10,5), width=1)
plt.title("Greens State/Territory MPs (single member electorates only)")
plt.ylabel("Number of MPs at year end")
#plt.axhline(y=33, color='grey', linestyle=':')
plt.savefig('state_mps_single member.png', dpi=300, bbox_inches='tight')
plt.show()
```
