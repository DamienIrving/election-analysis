```python
years = np.arange(1977, 2022)
countries = [
    'Australia',
    'Austria',
    'Belgium',
    'Brazil',
    'Bulgaria',
    'Canada',
    'Chile',
    'Cyprus',
    'New Zealand',
    'Rwanda',
]
```

```python
def fill_mps_list(mps_dict, years):
    """Create array of MPs for each year"""

    last_mps = 0
    mps_list = []
    for year in years:
        if year in mps_dict:
            last_mps = mps_dict[year]
        mps_list.append(last_mps)

    return np.array(mps_list)
```

```python
greens_mps_arrays = {}
total_mps_arrays = {}
for country in countries:
    greens_mps_arrays[country] = fill_mps_list(greens_mps[country], years)
    total_mps_arrays[country] = fill_mps_list(total_mps[country], years) 
```

```python
greens_df = pd.DataFrame(greens_mps_arrays)
greens_df = greens_df.set_index(years)
greens_df = greens_df[countries]
greens_df
```

```python
greens_df.plot(kind='bar', stacked=True, figsize=(9,5), width=1)
plt.title("Greens in National Parliaments")
plt.ylabel("Number of MPs at year end")
plt.savefig('global_greens_mps.png', dpi=300, bbox_inches='tight')
```

```python
total_df = pd.DataFrame(total_mps_arrays)
total_df = total_df.set_index(years)
total_df = total_df[countries]

greens_fraction_df = greens_df / total_df
```

```python
greens_fraction_df.plot(kind='bar', stacked=True, figsize=(9,5), width=1)
plt.title("Greens in National Parliaments")
plt.ylabel("Fraction of all MPs at year end")
plt.savefig('global_greens_parliament_fractions.png', dpi=300, bbox_inches='tight')
```
