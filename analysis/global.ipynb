```python
years = np.arange(1977, 2022)
countries = [
    'Australia',
    'Austria',
    'Belgium',
    'Brazil',
    'Bulgaria',
    'New Zealand',
]
```

```python
def fill_mps_list(mps_dict, years):
    """Create array of MPs for each year"""

    last_mps = 0
    mps_list = []
    for year in years:
        if year in mps_dict:
            last_mps = mps_dict[year]
        mps_list.append(last_mps)

    return np.array(mps_list)
```

```python
mps_arrays = {}
for country in countries:
    mps_arrays[country] = fill_mps_list(mps[country], years)
```

```python
df = pd.DataFrame(mps_arrays)
df = df.set_index(years)
df = df[countries]
df
```

```python
df.plot(kind='bar', stacked=True, figsize=(9,5), width=1)
plt.title("Greens in National Parliaments")
plt.ylabel("Number of MPs at year end")
plt.savefig('global.png', dpi=300, bbox_inches='tight')
```
