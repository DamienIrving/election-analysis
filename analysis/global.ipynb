```python
countries = list(greens_mps.keys())
countries
```

```python
ncountries = len(countries)
ncountries
```

```python
years = np.arange(1972, 2022)
```

```python
def fill_mps_list(mps_dict, years):
    """Create array of MPs for each year"""

    last_mps = 0
    mps_list = []
    for year in years:
        if year in mps_dict:
            last_mps = mps_dict[year]
        mps_list.append(last_mps)

    return np.array(mps_list)
```

```python
greens_mps_arrays = {}
total_mps_arrays = {}
for country in countries:
    greens_mps_arrays[country] = fill_mps_list(greens_mps[country], years)
    total_mps_arrays[country] = fill_mps_list(total_mps[country], years) 
```

```python
greens_df = pd.DataFrame(greens_mps_arrays)
greens_df = greens_df.set_index(years)
greens_df = greens_df[countries]
greens_df
```

```python
palette = sns.color_palette("tab20", 20).as_hex()
print(palette)
```

```python
tab40 = palette + palette
```

```python
greens_df.plot(kind='bar', stacked=True, figsize=(20,10), width=1, color=tab40[0:ncountries])
plt.title("Global Greens MPs in National Parliaments")
plt.ylabel("Number of MPs at year end")
plt.savefig('global_greens_mps.png', dpi=300, bbox_inches='tight')
# 'nipy_spectral'
```
