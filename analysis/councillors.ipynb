```python
count_dict = {'year': np.arange(1999, 2021)}
for council, member_dict in councillors.items():
    years = np.concatenate(list(member_dict.values()))
    counts, lower_bin_bounds = np.histogram(years, bins=np.arange(1998.5, 2021.5))
    count_dict[council] = counts
```

```python
df = pd.DataFrame(count_dict)
df = df.set_index('year')
df
```
