```python
# https://mokole.com/palette.html

mp_styles = {
    'Adam Bandt': ('#2f4f4f', 'o'),
    'Jamie Parker': ('#7f0000', 'o'),
    'Kobi Shetty': ('#7f0000', 's'),
    'Ellen Sandell': ('#008000', 'o'),
    'Sam Hibbins': ('#4b0082', 'o'),
    'Tamara Smith': ('#ff8c00', 'o'),
    'Jenny Leong': ('#ffff00', 'o'),
    'Jonathan Sri': ('#00ff00', 'o'),
    'Trina Massey': ('#00ff00', 's'),
    'Michael Berkman': ('#00ffff', 'o'),
    'Tim Read': ('#0000ff', 'o'),
    'Amy MacMahon': ('#ff00ff', 'o'),
    'Max Chandler-Mather': ('#6495ed', 'o'),
    'Stephen Bates': ('#ffdead', 'o'),
    'Elizabeth Watson-Brown': ('#ff69b4', 'o'),
}
```

```python
mp_list = [
    'Adam Bandt',
    'Jamie Parker',
    'Kobi Shetty',
    'Ellen Sandell',
    'Sam Hibbins',
    'Tamara Smith',
    'Jenny Leong',
    'Jonathan Sri',
    'Trina Massey',
    'Michael Berkman',
    'Tim Read',
    'Amy MacMahon',
    'Max Chandler-Mather',
    'Stephen Bates',
    'Elizabeth Watson-Brown',
]
plt.figure(figsize=[9, 5])
ax = plt.subplot(111)
for mp in mp_list:
    color, marker = mp_styles[mp]
    plt.plot(
        election_years[mp],
        primary_vote[mp],
        color=color,
        marker=marker,
        label=f"{mp} ({electorates[mp]})"
    )
plt.title('Greens sitting MP primary vote (single member electorates only)')
plt.ylabel('primary vote (%)')
plt.xlabel('election year')

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('greens_mps_singles.png', facecolor='white', bbox_inches='tight', dpi=300)
```
