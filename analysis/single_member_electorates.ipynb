```python
mp_list = ['Adam Bandt', 'Jamie Parker', 'Ellen Sandell',
           'Sam Hibbins', 'Tamara Smith', 'Jenny Leong',
           'Jonathan Sri', 'Michael Berkman', 'Tim Read']
plt.figure(figsize=[9, 5])
ax = plt.subplot(111)
for mp in mp_list:
    plt.plot(election_years[mp], primary_vote[mp],
             marker='o', label=f"{mp} ({electorates[mp]})")
plt.title('Greens MPs from (big) single member electorates (minimum two elections)')
plt.ylabel('primary vote (%)')
plt.xlabel('election year')

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('greens_mps_singles.png', facecolor='white', bbox_inches='tight', dpi=300)
```
