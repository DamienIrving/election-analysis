```python
mp_styles = {
    'Adam Bandt': ('tab:blue', 'o'),
    'Jamie Parker': ('tab:orange', 'o'),
    'Kobi Shetty': ('tab:orange', 's'),
    'Ellen Sandell': ('tab:green', 'o'),
    'Sam Hibbins': ('tab:red', 'o'),
    'Tamara Smith': ('tab:purple', 'o'),
    'Jenny Leong': ('tab:brown', 'o'),
    'Jonathan Sri': ('tab:pink', 'o'),
    'Trina Massey': ('tab:pink', 's'),
    'Michael Berkman': ('tab:gray', 'o'),
    'Tim Read': ('tab:olive', 'o'),
    'Amy MacMahon': ('tab:cyan', 'o'),
}
```

```python
mp_list = [
    'Adam Bandt',
    'Jamie Parker',
    'Kobi Shetty',
    'Ellen Sandell',
    'Sam Hibbins',
    'Tamara Smith',
    'Jenny Leong',
    'Jonathan Sri',
    'Trina Massey',
    'Michael Berkman',
    'Tim Read',
    'Amy MacMahon',
]
plt.figure(figsize=[9, 5])
ax = plt.subplot(111)
for mp in mp_list:
    color, marker = mp_styles[mp]
    plt.plot(
        election_years[mp],
        primary_vote[mp],
        color=color,
        marker=marker,
        label=f"{mp} ({electorates[mp]})"
    )
plt.title('Greens sitting MP primary vote (single member electorates only)')
plt.ylabel('primary vote (%)')
plt.xlabel('election year')

box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.savefig('greens_mps_singles.png', facecolor='white', bbox_inches='tight', dpi=300)
```
