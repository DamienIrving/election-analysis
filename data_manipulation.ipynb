```python
import pandas as pd
```

```python
votes_file = 'data/Lyons19-SenateDivisionFirstPrefsByPollingPlaceDownload-24310-196.csv'
votes_df = pd.read_csv(votes_file, skiprows=1)
```

```python
votes_df
```

```python
greens_votes_df = votes_df[votes_df.PartyNm == 'The Greens']
```

```python
greens_totals = greens_votes_df.groupby('PollingPlaceNm').OrdinaryVotes.sum()
greens_totals
```

```python
totals = votes_df.groupby('PollingPlaceNm').OrdinaryVotes.sum()
totals
```

```python
type(totals)
```

```python
totals_dict = {'total votes': totals.values,
               'greens votes': greens_totals.values,
               'greens percentage': (greens_totals.values / totals.values) * 100}
#'polling place': totals.index.values,
```

```python
pd.DataFrame(totals_dict, index=totals.index.values)
```
