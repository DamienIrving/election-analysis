```python
import pandas as pd
```

## Vote data

```python
votes_file = 'data/Lyons19-SenateDivisionFirstPrefsByPollingPlaceDownload-24310-196.csv'
raw_votes_df = pd.read_csv(votes_file, skiprows=1)
raw_votes_df
```

```python
greens_totals = raw_votes_df[raw_votes_df.PartyNm == 'The Greens'].groupby('PollingPlaceNm').OrdinaryVotes.sum()
greens_totals
```

```python
totals = raw_votes_df.groupby('PollingPlaceNm').OrdinaryVotes.sum()
```

```python
type(totals)
```

```python
totals_dict = {'PollingPlaceNm': totals.index.values,
               'total votes': totals.values,
               'greens votes': greens_totals.values,
               'greens percentage': (greens_totals.values / totals.values) * 100}
```

```python
votes_df = pd.DataFrame(totals_dict)
votes_df = votes_df.set_index('PollingPlaceNm')
votes_df
```

## Polling places

```python
polling_place_file = 'data/Fed19-GeneralPollingPlacesDownload-24310.csv'
raw_polling_place_df = pd.read_csv(polling_place_file, skiprows=1)
raw_polling_place_df
```

```python
polling_place_df = raw_polling_place_df[raw_polling_place_df['DivisionNm'] == 'Lyons'][['PollingPlaceNm', 'Latitude', 'Longitude']].set_index('PollingPlaceNm')
polling_place_df
```

## Join

```python
df = polling_place_df.join(votes_df).dropna()
df
```

```python
df.sort_values(by=['greens percentage'], ascending=False)
```
