```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_style("darkgrid")
```

```python
federal_mps = {}

federal_mps['WA'] = {"Jo Vallentine": np.arange(1990, 1992),
                     "Christabel Chamarette": np.arange(1992, 1996),
                     "Dee Margetts": np.arange(1993, 1999),
                     "Rachel Siewert": np.arange(2005, 2021),
                     "Scott Ludlam": np.arange(2008, 2017),
                     "Jordan Steele-John": np.arange(2017, 2021),
                    }

federal_mps['TAS'] = {"Bob Brown": np.arange(1996, 2012),
                      "Christine Milne": np.arange(2005, 2015),
                      "Nick McKim": np.arange(2015, 2021),
                      "Peter Whish-Wilson": np.arange(2012, 2021),
                     }

federal_mps['NSW'] = {"Michael Organ": np.arange(2002, 2004),
                      "Kerry Nettle": np.arange(2002, 2008),
                      "Lee Rhiannon": np.arange(2011, 2018),
                      "Mehreen Faruqi": np.arange(2018, 2021),
                     }

federal_mps['SA'] = {"Sarah Hanson-Young": np.arange(2008, 2021),
                     "Penny Wright": np.arange(2011, 2015),
                     "Robert Simms": np.arange(2015, 2016),
                    }

federal_mps['VIC'] = {"Adam Bandt": np.arange(2010, 2021),
                      "Richard Di Natale": np.arange(2011, 2020),
                      "Janet Rice": np.arange(2014, 2021),
                      "Lidia Thorpe": np.arange(2020, 2021),
                     }

federal_mps['QLD'] = {"Larissa Waters (term 1)": np.arange(2011, 2017),
                      "Larissa Waters (term 2)": np.arange(2018, 2021),
                      "Andrew Bartlett": np.arange(2017, 2018),
                     }
```

```python
mp_count_dict = {'year': np.arange(1990, 2021)}
for state, mp_dict in federal_mps.items():
    years = np.concatenate(list(mp_dict.values()))
    counts, lower_bin_bounds = np.histogram(years, bins=np.arange(1989.5, 2021.5))
    mp_count_dict[state] = counts
```

```python
df = pd.DataFrame(mp_count_dict)
df = df.set_index('year')
df
```

```python
df.plot(kind='bar', stacked=True, figsize=(9,5), width=1)
plt.title("Greens Federal MPs")
plt.ylabel("Number of MPs at year end")
plt.savefig('federal_mps.png', dpi=300, bbox_inches='tight')
```
