```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dates = pd.to_datetime([
    '2024-02-17',
    '2024-03-23',
    '2024-05-17',
    '2024-08-17',
    '2024-11-17',
    '2025-02-17',
    '2025-05-17',
    '2025-06-17',
    '2025-07-01',
    '2025-07-08',
    '2025-07-15'
])
```

```python
data = {
    'Liberal': [39, 37, 35, 36, 35, 34, 29, 32, 35, 37, 37],
    'Labor': [26, 29, 28, 27, 31, 30, 31, 29, 28, 26, 26],
    'Greens': [12, 14, 15, 14, 14, 13, 14, 14, 14, 15, 14],
    'Independent': [14, 10, 12, 12, 11, 12, 17, 19, 18, 18, 19],
    'Jacqui Lambie Network': [9, 7, 7, 8, 6, 8, 6, np.nan, np.nan, np.nan, np.nan],
    'Nationals': [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, np.nan, 2, 2, 3, np.nan],
    'Other': [1, 4, 3, 3, 3, 3, 4, 4, 4, np.nan, np.nan],
}
```

```python
df = pd.DataFrame(data=data, index=dates)
```

```python
df.sum(axis=1)
```

```python
ax = df.plot(
    color=['tab:blue', 'tab:red', 'limegreen', 'tab:gray', '#F1CE63', 'tab:green', 'tab:pink'],
    marker='o',
    figsize=[8, 5],
)
ax.grid()
ax.axhline(y=37, color='tab:blue', linestyle=':')
ax.axhline(y=29, color='tab:red', linestyle=':')
ax.axhline(y=14, color='limegreen', linestyle=':')
ax.axhline(y=10, color='tab:gray', linestyle=':')
ax.axhline(y=4, color='tab:pink', linestyle=':')
ax.axvline(x=dates[1], color='black', linestyle=':', label='last election')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.set_title('EMRS polling (Tasmania, statewide)')
ax.set_ylabel('primary vote (%)')
plt.savefig('emrs.png', bbox_inches='tight', facecolor='white', dpi=300)
```
