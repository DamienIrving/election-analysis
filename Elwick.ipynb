```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=[14, 6], sharey=True,
                               gridspec_kw={'width_ratios': [2, 3]})

xvals2018 = np.array([1, 2, 3, 4])
yvals2018 = np.array([48.7, 36.3, 11.6, 3.4])
ax1.bar(xvals2018, yvals2018, color=['tab:red', 'tab:blue', 'tab:green', 'tab:grey'])
ax1.set_xticks(xvals2018)
ax1.set_xticklabels(['Labor', 'Liberal', 'Greens', 'Other'])
ax1.set_title('2018 State Election')
ax1.set_ylabel('Primary vote (%)')
ax1.grid(axis='y')

xvals2021 = np.array([1, 2, 3, 4, 5, 6])
yvals2021 = np.array([28.5, 29.0, 15.2, 13.2, 8.5, 5.6])
ax2.bar(xvals2021, yvals2021, color=['tab:red', 'tab:blue', 'tab:green', 'tab:cyan', 'tab:purple', 'tab:grey'])
ax2.set_xticks(xvals2021)
ax2.set_xticklabels(['Labor', 'Liberal', 'Greens', 'Johnston', 'Hickey', 'Other'])
ax2.set_title('2021 State Election')
ax2.grid(axis='y')

plt.suptitle('Elwick polling places')
plt.savefig('elwick.png', dpi=300)
```
